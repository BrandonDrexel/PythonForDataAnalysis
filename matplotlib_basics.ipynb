```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
job_titles = ['Data Analyst', 'Data Scientist', 'Data Engineer']
postings = [50, 75, 60]

plt.bar(job_titles, postings)
```

```python
months = ['January', 'February', 'March', 'April']
postings = [120, 150, 100, 130]

plt.plot(months, postings)
```

```python
# Create a bar chart
# showing the number of job postings for the job_country.
# Use the job_country column and count the number of postings for each location.
# Only get the top 5 job countries.
# Additionally, change the color of the bars to 'orange'.

postings_per_country = df.job_country.value_counts().head()
plt.bar(postings_per_country.index, postings_per_country, color='orange')
plt.bar(postings_per_country.index, postings_per_country.values, color='orange')
plt.xlabel('Job Country')
plt.ylabel('Number of Postings')
plt.title('Numbers of Jobs per Country')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Create a horizontal bar chart
# showing the average yearly salary for different job titles.
# Use the job_title_short and salary_year_avg columns to calculate the average salary for each job title.

average_yearly_salary = df.groupby('job_title_short')['salary_year_avg'].mean()
average_yearly_salary.sort_values(ascending=True, inplace=True)
plt.barh(average_yearly_salary.index, average_yearly_salary.values)
plt.xlabel('Average Salary')
plt.title('Average Salary by Job Title')
```

```python
# Create a line plot
# showing the trend of average yearly salary over the months of the year.
# Use the job_posted_date column to extract the month and calculate the average salary for each month.

year_to_month = df.job_posted_date.dt.month
average_monthly = df.groupby(year_to_month)['salary_year_avg'].mean()
plt.plot(average_monthly.index, average_monthly.values, color='red')
plt.xlabel('Month')
plt.ylabel('Average Salary')
plt.title('Average Salary by Job Month')
plt.legend(labels=['Average Salary'])
plt.show()
```

```python
# Create a bar plot
# with the Pandas .plot() method
# showing the top 5 job posting platforms (e.g., LinkedIn, Upwork, etc.) with the highest number of job postings.
# Use the job_via column to count the number of postings for each platform.

top_five_platforms = df.job_via.value_counts().head()
top_five_platforms.plot(kind='bar')
```

```python
# Create a line plot showing the trend of job postings over time using the Pandas method to plot:
# Create a DataFrame called monthly_job_counts and use the job_posted_date column to count the number of job postings for each month (in the format 'Jan').
# Add labels for the x-axis (label as 'Month'), y-axis (label as 'Number of Job Postings'), and a title (set as 'Trend of Job Postings Over Time') for the chart.

df.job_posted_month = df.job_posted_date.dt.to_period('M')
df.job_posted_month.value_counts().sort_index()
monthly_job_counts.plot(kind='line')
plt.xlabel('Month')
plt.ylabel('Number of Job Postings')
plt.title('Trend of Job Postings Over Time')
plt.show()
```
