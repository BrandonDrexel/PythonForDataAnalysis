```python
#Check for missing values in the job_skills column and display the rows where the values are not NA.
not_na = df['job_skills'].notna()
df[not_na]
```

```python
# Sort the DataFrame by salary_year_avg in descending order and display the first 10 rows.

descending = df.sort_values(by='salary_year_avg', ascending=False)
descending.head(10)
```

```python
# Create a new column is_remote that indicates whether the job is remote (True) or not (False) based on the job_work_from_home column.

df['is_remote'] = df['job_work_from_home']
df.head()
```

```python
# Create a new column job_posted_year that extracts the year from the job_posted_date column and display the first 5 rows.

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_posted_year'] = df['job_posted_date'].dt.year
df.head()
```

```python
# Drop rows where the salary_year_avg column has missing values and display the first 5 rows of the cleaned DataFrame.

droppies = df.dropna(subset='salary_year_avg')
droppies.head()
```

```python
df['salary_year_avg'].mean()
```

```python
df['salary_year_avg'].median()
```

```python
# Group the DataFrame by job_title_short and calculate the average salary_hour_avg for each job title. Display the result.

df.groupby('job_title_short')['salary_hour_avg'].mean()
```

```python
# Find the job title with the lowest average hourly salary and display the job title and the salary.

minimum_index = df['salary_hour_avg'].idxmin()
df.iloc[minimum_index][['job_title_short', 'salary_hour_avg']]
```

```python
# Group the DataFrame by job_country and calculate the total number of job postings for each country. Display the result.

df.groupby('job_country').size()
```
